### Part 1: Data Preparation

```python
df.info()
```

```python
df.describe().transpose()
```

```python
df = df.drop(501,axis=0)
```

```python
df = df.drop(df[df["course_name"].isnull()].index,axis=0)
```

```python
df.isnull().sum()
```

```python
df["course_flag"] = df["course_flag"].fillna("None")
```

```python
df = df.dropna()
```

```python
df["main_price"] = df["main_price"].apply(lambda x: float(x.split("£")[1].replace(",",""))*0.039)
```

```python
df["reviews_avg"] = df["reviews_avg"].apply(lambda x: float(x.split(" ")[1]))
```

```python
df["reviews_count"]= df["reviews_count"].apply(lambda x: int(x.split(" ")[0]))
```

```python
df["course_duration"] = df["course_duration"].apply(lambda x: float(x.split(" ")[0]))
```

```python
df["lectures_count"] = df["lectures_count"].apply(lambda x: int(x.split(" ")[0]))
```

```python
df["students_count"] = df["students_count"].apply(lambda x: int(x.split(" ")[0].replace(",","")))
```

## Part 2: EDA 

```python
df.describe().transpose()
```

```python
sns.countplot(data=df,x="course_flag")
```

#### Are longer courses more expensive?

```python
plt.figure(figsize=(12,6),dpi=150)
sns.scatterplot(data=df,x="course_duration",y="main_price")
```

### Does courses with higher review rate cost more?

```python
plt.figure(figsize=(12,6),dpi=150)
sns.boxplot(data=df,x="course_flag",y="reviews_avg")
```

### Do courses with more reviews have higher review rate?

```python
plt.figure(figsize=(12,6),dpi=150)
sns.scatterplot(data=df,x="reviews_avg",y="reviews_count",hue="course_flag")
```

### Do instructors with that have more students also create more lectures?

```python
plt.figure(figsize=(12,6),dpi=150)
sns.scatterplot(data=df,y="students_count",x="lectures_count")
```
